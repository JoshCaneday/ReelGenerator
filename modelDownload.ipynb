```python
from huggingface_hub import snapshot_download

local_dir = "./models/XTTS-v2"

snapshot_download(
    repo_id="coqui/XTTS-v2",
    revision="main",          # or pin e.g. "v2.0.2"
    local_dir=local_dir,
    local_dir_use_symlinks=False,
)
print("downloaded to", local_dir)
```

```python
from huggingface_hub import snapshot_download
from pathlib import Path

local_dir = "./models/CLIP-RN50"

snapshot_download(
    repo_id="jinaai/clip-models",
    revision="main",                 # optionally pin to a commit hash for reproducibility
    local_dir=local_dir,
    local_dir_use_symlinks=False,
    allow_patterns=["RN50.pt"],       # only grab what you need
)

model_path = str(Path(local_dir) / "RN50.pt")
print("downloaded to", model_path)
```
